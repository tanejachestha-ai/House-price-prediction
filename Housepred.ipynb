1Ô∏è‚É£ Why Encoding is Needed?
Explain briefly that ML models require numerical input,so categorical variables must be converted into numbers.

2Ô∏è‚É£ How Encoding Was Done?
Label Encoding For binary or ordinal features (like size). 
Target Guided Encoding: For nominal features with multiple categories (like area_type,location,society).

```python
area_price=df.groupby("area_type")["price"].mean().to_dict()
```

```python
area_price
```

```python
df["new_Area"]=df["area_type"].map(area_price)
df
```

Insight:
We applied target-guided encoding on the area_type column to create a new numerical feature new_Area. This replaces categorical labels with values reflecting their average price impact, allowing the model to learn meaningful patterns from area_type instead of treating it as an arbitrary category.

```python
x=df.drop(columns=["price"])
y=df["price"]
```

Insight:
We separated the dataset into features (x) and target (y), where x contains all predictor variables and y contains the property price. This prepares the data for modeling and machine learning algorithms.

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
```

```python
x=scale.fit_transform(x)
x
```

Insight:
We applied StandardScaler to scale the feature values in x. This ensures all features have similar ranges and zero mean, which helps models converge faster and perform better, especially for algorithms sensitive to feature magnitude.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=0)
```

Insight:
We split the dataset into training and testing sets using an 80:20 ratio. This allows the model to learn patterns from the training data and be evaluated on unseen test data to assess its performance and generalization ability.

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor=RandomForestRegressor()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

Insight:
We trained a Random Forest Regressor to predict property prices. This ensemble model combines multiple decision trees, improving accuracy and reducing overfitting. It can handle both numerical and categorical features effectively and provides robust predictions on the test set.

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(y_test,y_pred))
```

Insight:
The Random Forest Regressor achieved an R¬≤ score of 88%, indicating that the model explains 88% of the variance in property prices. This demonstrates that the model is highly effective and reliable for predicting real estate prices in the dataset.

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

Insight:
We trained a Linear Regression model on the house price dataset to predict property prices. This model captures the linear relationships between features and the target, providing a simple yet effective baseline with good predictive performance.

```python
y_pred
```
